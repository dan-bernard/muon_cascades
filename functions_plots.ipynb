```python
import matplotlib.pyplot as plt
import math
```

```python
def n(Z):
    return math.sqrt((Z*(1+2*math.sqrt(Z)))/(1+2/math.sqrt(Z)))
```

```python
ListZ = [Z for Z in range(1,100)]
Listn = [n(Z) for Z in ListZ]
```

```python
plt.plot(ListZ, Listn)
plt.xlabel('Z')
plt.ylabel('n')
plt.show()
```

```python
def ComputeCSq(n,l):
    if l == 0:
        return 1/n
    else:
        CPrevious = ComputeCSq(n,l-1)
        return CPrevious * (2*l+1)*(n-l)/((2*l-1)*(n+l))
```

```python
def Factorial(n):
    if n == 0:
        return 1
    else:
        return n*Factorial(n-1)
```

```python
def CSq(n,l):
    if l >= n:
        return 0
    return (2*l+1)*(Factorial(n-1)**2)/Factorial(n-l-1)/Factorial(n+l)
```

```python
Atoms = {"Mg":12,"Al":13,"Fe":26,"In":49,"Ho":67,"Au":79}
Ls = [ l for l in range(19) ]
```

```python
plt.xlabel("l")
plt.ylabel("C(n,l)")

for atom in Atoms:
    Cls = [ ComputeCSq(int(n(Atoms[atom])),l) for l in Ls ]
    plt.plot(Ls,Cls,label=atom+", Z = "+str(Atoms[atom]))

plt.legend(loc="best")
plt.show()
```
